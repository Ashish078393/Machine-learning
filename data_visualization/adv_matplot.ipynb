```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.csv')
iris.sample(4)
```

```python
pd.set_option('future.no_silent_downcasting',True)
iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2},inplace=True)
iris.sample(4)
```

```python
plt.figure(figsize=(13,7)) #figsize=(widht,lenght)
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.44) #here 'cmap' is color map to map colors .
plt.xlabel('Sepal lenght')
plt.ylabel('Petal lenght')
plt.colorbar()
```

***Annotation***

```python
batters = pd.read_csv('batter.csv')
batters.head()
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
sample_df
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2,6,'Point 2')
plt.text(3,7,'Point 3')
plt.text(4,8,'Point 4',fontdict={'size':12,'color':'red'})
plt.show()
```

```python
plt.figure(figsize=(18,12))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(18,12))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color='red')
plt.axvline(30,color='red')
plt.axhline(140,color='green')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
fig,ax = plt.subplots()
ax.scatter(batters['avg'],batters['strike_rate'])
ax.set_title('something scatter plot')
ax.set_xlabel('Avg')
ax.set_ylabel('strike_score')
plt.show()
```

```python
fig,ax = plt.subplots(2,2,figsize=(12,6))
ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')
ax2.hist(batters['runs'])
plt.show()
```

```python
x = [0,10,5]
y = [0,10,13]
z = [0,13,20]
fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter3D(x,y,z,s=[100,100,100])
ax.plot3D(x,y,z,color='red')
ax.set_xlabel('runs')
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
XX,YY = np.meshgrid(x,y)
XX.shape
```

```python
z = XX**2+YY**2
z.shape
```

```python
fig =plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(XX,YY,z,cmap='viridis')
plt.colorbar(p)
```

***contour plots***

```python
fig =plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(XX,YY,z,cmap='viridis')
plt.colorbar(p)
```

```python
#filled contour
fig =plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(XX,YY,z,cmap='viridis')
plt.colorbar(p)
```

```python
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022 (1).csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6]))&(delivery['batsman_run'] == 6)]
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
grid.head()
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.colorbar()
```
