```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='--',linewidth=2,marker='D',markersize=8,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='green',linestyle='solid',linewidth=3,marker='+',markersize=11,label='rohit')
plt.legend(loc='upper left')
plt.title('virat and Rohit runs')
plt.xlabel('yaer')
plt.ylabel('runs')
plt.grid()
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
children = [10,20,40,10,30]
colors = ['red red red','blue blue blue','green green green','yellow yellow yellow','pink pink pink']
plt.bar(colors,children)
plt.xticks(rotation='vertical')
```

```python
#stack bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016']+df['2017']),label='2015')
plt.legend()
plt.show()
```
