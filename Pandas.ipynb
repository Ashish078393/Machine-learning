```python
df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['keerti','Amit','Educosys','Youtube']
})
```

```python
df2 = pd.DataFrame({
    'id':[3,4,5,6],
    'courses':['GenAI','CPP','LLD','HLD']
})
```

```python
inner_merge = pd.merge(df1,df2,how='inner',on='id')
inner_merge
```

```python
outer_merge = pd.merge(df1,df2,how='outer',on='id')
outer_merge
```

```python
left_result = pd.merge(df1,df2,how='left',on='id')#it will not give lef side none value
left_result
```

```python
right_result = pd.merge(df1,df2,how='right',on='id')#it will not give right side none value
right_result
```
